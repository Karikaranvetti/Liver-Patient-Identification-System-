## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Indian Liver Patient Dataset (ILPD).csv', header=None)
m = dataset[9].mean()
dataset[9].fillna(m, inplace = True) 
print(dataset.to_string()) 

 
```

```python
dataset.info()
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
type(X)
```

## Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = ct.fit_transform(X)
type(X)
```

```python
X[1,:]
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X[:,2:], y, test_size = 0.2, random_state = 1)
type(X_train)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)     # here we are fit the value for train  set and transform 
X_test= sc.transform(X_test)    # here we just transform with train data value because test data set is actualy supries one 
```

## Training the Logistic Regression model on the Training set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0 ).fit(X_train, y_train)
# classifier
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

## Making the Confusion Matrix  Training data accuracy_score

```python
y_pred1 = classifier.predict(X_train)
cm = confusion_matrix(y_train, y_pred1)
print(cm)
accuracy_score(y_train, y_pred1)
```
